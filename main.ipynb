```python
x = house_data[['Area', 'NoOfBedrooms', 'Resale', 'LiftAvailable', 'CarParking']]
y = house_data[['Price']]/100000

print(x)
print(y)
```

Jag skapar två olika variblar x och y, där x får innehålla de kolumnerna som jag tycker kan vara mest intressanta för denna labb. Bland annat de tre kolumnerna som hade mest korrelation till priset och sen två övriga som kan vara för grund till slutpriset. Variabeln y får innehålla priset och jag skalar ner priset för att det ska gå snabbare och enklare att hantera datan.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7)
```

importerar train_test_split för att kunna dela upp datan i olika train och test sets. 

Jag delar sedan upp x och y i varsin 'train' och 'test' datasets. 
X och y inne i parantesen innehåller data som jag specificerade högre upp.

x --> Area, NoOfBedrooms, Resale ..... 

y --> Price 

Därefter så specificerar jag att 70% av datan ska vara i respektive x_ och y_train set och resterande 30% i x- och y_test setten.

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Kör en shape på vardera train och test set. I outputen så ser vi ju även att båda x- och y-train seten har fått ca 70% data i sig och x_ och y_test seten har fått ca 30% av datan, precis som jag planerat.

```python
from sklearn.linear_model import LinearRegression

lin_regr = LinearRegression().fit(x_train, y_train)
y_test_pred = lin_regr.predict(x_test)
```

Här har jag importerat LinearRegression och använder mig av fit() för att kunna anpassa modellen med hjälp av x_ och y_train. Vad som menas är att den tränar modellen baserat på x_ och y_train värdena. X_train innehåller ju alla egenskaper i huset från datasetet och y_train innehåller priskolumnen. 
Efter att modellen har "tränats" så vill vi nu förutse värdet för x_test. Det har jag gjort genom att använda modellen som jag precis byggt, "lin_regr", som sedan lagrar sina värden i varibeln "y_test_pred" och vilka värden modellen än förutser för x_test så kommer det också att lagaras i varibeln "y_test_pred".

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(16,9))
plt.scatter(range(len(y_test[:150])), y_test[:150], color='black', label='Test data')
plt.scatter(range(len(y_test_pred[:150])), y_test_pred[:150], color='darkorange', marker='*', label='Förutsagd data')
plt.xlabel('Hus')
plt.ylabel('Hus Priser')
plt.title('Test data vs förutsagd data')
plt.legend()
plt.show()
```

Här skapar jag en visualisering av sluttresultatet med hjälp av pyplot från matplotlib. 

Där de svarta prickarna är testdatan vilket är ursprungsvärdet och de oranga stjärnorna prickarna är den uträknade datans värden.
